```python
df['Storage '] = df['Storage '].str.replace(' ', '')
df['Storage '].unique()
```

```python
print(df['Storage '].dtype)
```

Analisis kolom RAM

```python
df['RAM '].unique()
```

```python
df['RAM '] = df['RAM '].str.replace(' ', '')
df['RAM '] = df['RAM '].str.replace('GB', '')
```

```python
df['RAM '] = df['RAM '].astype(int)
```

```python
df.describe()
```

Analisis Price

```python
df['Price ($)'].unique()
```

```python
df['Price ($)'] = df['Price ($)'].str.replace(' ', '')
df['Price ($)'] = df['Price ($)'].str.replace('$', '')
df['Price ($)'] = df['Price ($)'].str.replace(',', '')
df['Price ($)'] = df['Price ($)'].astype(int)
df['Price ($)'].dtype
```

```python
df.describe()
```

Filter

```python
df.loc[df['Brand']=='Xiaomi'].sort_values(by = 'Price ($)', ascending = True)
```

```python
#memfilter data berdasarkan brand dengan kondisi harga lebih dari 190 dan kurang dari 900 dengan tampilan kolom Brand, Model, dan harga kemudian diurutkan dari terkecil ke terbesar
df.loc[(df['Brand'] == 'Xiaomi') & (df['Price ($)'] > 190) & (df['Price ($)'] < 900), ['Brand', 'Model', 'Price ($)']]
df.sort_values(by = 'Price ($)', ascending = True)
```

```python
df.iloc[2:10,0:4]
```

```python
df['Battery dollar'] = df['Battery Capacity (mAh)'] / df['Price ($)']
```

```python
df.sort_values(by = 'Battery dollar', ascending = False)
```
